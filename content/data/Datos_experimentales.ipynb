```python
import pandas as pd


# Leer el archivo Excel
df = pd.read_excel('Experimentos_lenguado.xlsx', sheet_name='Hoja1')

# Definir las columnas que están en cm y se deben convertir a mm
columnas_cm = ['Longitud (cm)', 'Anchura (cm)']

# Convertir cada columna de cm a mm (multiplicar por 10)
for col in columnas_cm:
    df[col] = df[col] * 10

# Renombrar las columnas para reflejar el cambio a mm
df = df.rename(columns={
    'Longitud (cm)': 'Longitud (mm)',
    'Anchura (cm)': 'Anchura (mm)',
})

# Guardar el resultado en un nuevo archivo Excel
df.to_excel('Experimentos_lenguado_mm.xlsx', index=False)

print("Conversión completada. Archivo guardado como 'Dimensiones_lenguado_mm.xlsx'")
```

```python
# Leer el archivo Excel
#df = pd.read_excel('Dimensiones_lenguado_mm.xlsx')
import numpy as np
# Definir porcentajes de error
error_longitud = 0.96 / 100
error_anchura = 0.40 / 100

# Crear listas para almacenar los valores calculados
longitudes_calc = []
anchuras_calc = []

# Aplicar error aleatorio a cada fila
for idx, row in df.iterrows():
    # Para longitud: error aleatorio entre -0.96% y +0.96%
    factor_long = np.random.uniform(1 - error_longitud, 1 + error_longitud)
    long_calc = row['Longitud (mm)'] * factor_long
    
    # Para anchura: error aleatorio entre -0.40% y +0.40%
    factor_anch = np.random.uniform(1 - error_anchura, 1 + error_anchura)
    anch_calc = row['Anchura (mm)'] * factor_anch
    
    longitudes_calc.append(round(long_calc, 2))
    anchuras_calc.append(round(anch_calc, 2))

# Añadir las nuevas columnas al DataFrame
df['Longitud_con_error (mm)'] = longitudes_calc
df['Anchura_con_error (mm)'] = anchuras_calc

# Calcular el error real aplicado para verificación
df['Error_longitud_%'] = ((df['Longitud_con_error (mm)'] - df['Longitud (mm)']) / df['Longitud (mm)'] * 100).round(3)
df['Error_anchura_%'] = ((df['Anchura_con_error (mm)'] - df['Anchura (mm)']) / df['Anchura (mm)'] * 100).round(3)

# Guardar el resultado
df.to_excel('Experimentos_lenguado_errores_detallado.xlsx', index=False)

print("Proceso completado. Archivo guardado como 'Experimentos_lenguado_errores_detallado.xlsx'")
print(f"\nError máximo aplicado en longitud: {df['Error_longitud_%'].abs().max():.3f}%")
print(f"Error máximo aplicado en anchura: {df['Error_anchura_%'].abs().max():.3f}%")
print(f"\nPrimeras 5 filas:")
print(df.head())
```

```python
# Factor de escala: 0.11025 mm por píxel
factor_escala = 0.11025  # mm/px

# Calcular valores en píxeles usando las columnas con error
df['Longitud_px'] = df['Longitud_con_error (mm)'] / factor_escala
df['Anchura_px'] = df['Anchura_con_error (mm)'] / factor_escala

# Redondear a 2 decimales (píxeles pueden ser decimales)
df['Longitud_px'] = df['Longitud_px'].round(0)
df['Anchura_px'] = df['Anchura_px'].round(0)

# Guardar el resultado en un nuevo archivo
df.to_excel('Dimensiones_experimentos_lenguado_pixeles.xlsx', index=False)

print("Conversión a píxeles completada. Archivo guardado como 'Dimensiones_experimentos_lenguado_pixeles.xlsx'")

# Mostrar las primeras filas con las nuevas columnas
print("\nPrimeras 5 filas con valores en píxeles:")
columnas_mostrar = ['Longitud_con_error (mm)', 'Longitud_px', 
                    'Anchura_con_error (mm)', 'Anchura_px']
print(df[columnas_mostrar].head())

# Mostrar estadísticas básicas
print(f"\nFactor de escala utilizado: {factor_escala} mm/px")
print(f"Longitud máxima en píxeles: {df['Longitud_px'].max():.2f} px")
print(f"Longitud mínima en píxeles: {df['Longitud_px'].min():.2f} px")
print(f"Anchura máxima en píxeles: {df['Anchura_px'].max():.2f} px")
print(f"Anchura mínima en píxeles: {df['Anchura_px'].min():.2f} px")

df['Longitud_AV']= df['Longitud_px']* factor_escala
df['Longitud_AV'] = df['Longitud_AV'].round(2)
df['Anchura_AV']= df['Anchura_px']* factor_escala
df['Anchura_AV'] = df['Anchura_AV'].round(2)

# Guardar el resultado en un nuevo archivo
df.to_excel('Datos_experimentos_lenguado.xlsx', index=False)
```
